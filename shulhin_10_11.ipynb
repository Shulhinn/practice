### **Завдання 1 (2 бали)**

Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```python
def factorial(x):
    if x == 0:
        return 1
    fac = x
    for i in range(1, x):
        fac *= i
    return fac

def binomial(n):
    for x in range(n + 1):
        coefs = []
        for y in range(x + 1):
            x_fac = factorial(x)
            y_fac = factorial(y)
            xy_fac = factorial(x - y)
            coefs.append(x_fac // y_fac // xy_fac)
        yield coefs

n = int(input())
for binomials in binomial(n):
    print(' '.join([str(b) for b in binomials]))
```

У нашій книзі таких великих листів буде 4. Розглянемо перший лист. З одного боку на ньому будуть надруковані сторінки 16 та 1, з іншого - 2 та 15. Якщо параметр дорівнює False, додаткові функції декоратора не повинні виконуватись, тобто має виконатись лише функція з п. 4. Декоратор повинен створювати список усіх сторінок у потрібній послідовності, причому номери сторінок кожного зошита мають зберігатись в окремому вкладеному списку. Для 16-сторінкового зошита список повинен мати вигляд:

**Бонусне завдання (за бажанням)**

6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```python
book_pages_number = int(input('Введіть кількість сторінок у книзі (не більше 1280): '))
print(book_pages_number)
if book_pages_number > 1280:
    print('Неправильна кількість сторінок! У книзі може бути не більше 1280 сторінок!')
    exit(1)

notebook_pages_number = int(input('Введіть кількість сторінок у зошиті (дозволені значення: 16, 24 та 32): '))
print(notebook_pages_number)
if notebook_pages_number not in [16, 24, 32]:
    print('Неправильна кількість сторінок у зошиті! У зошиті повинно бути 16, 24 або 32 сторінки!')
    exit(1)

if book_pages_number % notebook_pages_number != 0:
    print('Кількість сторінок у книзі повинна бути кратною кількості сторінок у зошиті!')
    exit(1)

notebooks_in_book_number = book_pages_number // notebook_pages_number

with_extra = int(input('Введіть 0 або 1'))
if with_extra not in [0, 1]:
    print('Треба ввести 0 або 1!')
    exit(1)

def by_notebook(extra):
    def with_extra(f):
        if extra == 1:
            def print_by_notebook():
                notebooks = f()
                for nb in notebooks:
                    pages = []
                    for j in range(len(nb) // 2):
                        if j % 2 == 0:
                            pages.append(nb[len(nb) - j - 1])
                            pages.append(nb[j])
                        else:
                            pages.append(nb[j])
                            pages.append(nb[len(nb) - j - 1])
                    yield pages
            return print_by_notebook
        else:
            return f
    return with_extra

@by_notebook(with_extra)
def print_pages():
    for i in range(notebooks_in_book_number):
        offset = i * notebook_pages_number
        pages = [offset + x for x in range(1, notebook_pages_number + 1)]
        yield pages

print('Зошитів у книзі: {}'.format(notebooks_in_book_number))
for nb in print_pages():
    print(nb)
```
